## Control Flow (if statements)

```python
if var1 or var2:
  print("benar")
else:
  print("salah")
```

```python
if var1 and var2:
  print("benar")
elif var1 > var2:
  print("ini kondisi kedua")
else:
  print("salah")
```

```python
if var1 or var2:
  if var1 > var2:
    print("ini kondisi kedua")
else:
  print("salah")
```
